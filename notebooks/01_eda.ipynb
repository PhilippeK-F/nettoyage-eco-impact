Cellule 1 - Imports et chargements

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Style global des graphiques
sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams["figure.figsize"] = (10, 5)

df = pd.read_csv("../data/produits_clean.csv", parse_dates=["date"])
print(f"Lignes : {len(df)} | Colonnes : {len(df.columns)}")
df.head()
```

Cellule 2 - Vue d'ensemble 

```python
# Résumé statistique des indicateurs écologiques
cols_eco = [
    "cov_g_par_l", "dbo5_mg_par_l", "biodegradabilite_pct",
    "lc50_daphnie_mg_par_l", "score_env"
]
df[cols_eco].describe().round(2)
```

Cellule 3 - Distribution du score environnemental 

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Histogramme
axes[0].hist(df["score_env"], bins=20, color="#4C9BE8", edgecolor="white")
axes[0].set_title("Distribution du score environnemental")
axes[0].set_xlabel("Score (0–10)")
axes[0].set_ylabel("Nombre de produits")

# Boxplot par type chimique
order = df.groupby("type_chimique")["score_env"].median().sort_values().index
sns.boxplot(
    data=df, x="score_env", y="type_chimique",
    order=order, ax=axes[1], palette="coolwarm"
)
axes[1].set_title("Score par type chimique")
axes[1].set_xlabel("Score environnemental")
axes[1].set_ylabel("")

plt.tight_layout()
plt.savefig("../reports/01_distribution_score.png", dpi=150)
plt.show()
```

Cellule 4 - COV par catégorie

```python
cov_cat = (
    df.groupby("categorie")["cov_g_par_l"]
    .mean()
    .sort_values(ascending=False)
)

plt.figure(figsize=(10, 5))
bars = plt.bar(cov_cat.index, cov_cat.values, color="#E8724C", edgecolor="white")
plt.xticks(rotation=35, ha="right")
plt.title("Émission COV moyenne par catégorie de produit (g/L)")
plt.ylabel("COV moyen (g/L)")
plt.tight_layout()
plt.savefig("../reports/02_cov_par_categorie.png", dpi=150)
plt.show()
```

Cellule 5 - Matrice de corrélation

```python
cols_corr = [
    "score_env", "biodegradabilite_pct", "lc50_daphnie_mg_par_l",
    "cov_g_par_l", "ph", "prix_litre_eur"
]

corr = df[cols_corr].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(
    corr, annot=True, fmt=".2f", cmap="RdYlGn",
    center=0, square=True, linewidths=0.5
)
plt.title("Corrélations entre indicateurs")
plt.tight_layout()
plt.savefig("../reports/03_correlations.png", dpi=150)
plt.show()
```
